# E-Commerce Customer Analysis (Python + SQL)

## Overview
Analyzed 500k+ retail transactions to identify revenue trends, top products/countries, and customer segments using Python (pandas) and SQL (SQLite).

## Dataset
Kaggle Online Retail dataset (transaction-level purchases).

## Tools
- Python: pandas, matplotlib
- SQL: SQLite
- Jupyter Notebook

## Key Work
- Cleaned and validated 541k+ rows (removed duplicates, missing CustomerID rows, returns)
- Feature engineering: revenue, year/month
- KPI reporting: total revenue, customers, AOV, monthly trend
- Customer analytics: RFM segmentation (Champions, At Risk, Lost, etc.)
- SQL portfolio queries for business reporting
